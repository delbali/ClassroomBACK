## Assignment operator

Let's have a better look at the line:

`A=B`

The operator "=" is called __assignment operator__, and it works from right to left: first evaluates B and then assigns it the name A.

For example:

````python
a = 3 + 2
print(a)
````

In Python, variables don’t store data directly but rather store references to objects in memory.

When you assign a number to a variable (like an integer or a float), Python treats the variable as a label for that specific value. If you assign a new variable to the first one (y = x), Python creates a copy of the value.

Because numbers are immutable (they cannot be changed, only replaced), x and y act as independent variables.

````python
x = 5
y = x # y gets the value 5
x = 10 # x is now pointing to a new value (10)

print(x)
print(y)
````

When you assign a variable to a list, the variable does not hold the list itself; it holds its reference, the place where that list lives in the computer's memory.

````python
a = [1, 2, 3]
b = a # b now points to the SAME list as a
a.append(4) # We modify the list via a

print("a is equal to ", a)
print("b is equal to ", b)
````

## Test 2: some geometry (cylinder.py)

* Write a code to compute the volume and total surface area of a cylinder of radius 4 and height 6.
* EXTRA (1): modify the previous code to take user input for the radius value.
* EXTRA (2): make sure input and outputs are easy to read and understand.


Some help in case you forgot the details:

  $S = 2πr(r + h)$
  
  $V = πr^2*h$

````python
import math #we import the math module, needed for the math.pi variable (better then approximating)

height = 6

# EXTRA (1): Taking user input for the radius
try:
    user_input = input("Enter the radius of the cylinder: ")
    radius = float(user_input)
except ValueError:
    print("Invalid input! Please enter a numeric value.")
    # Defaulting to 4 if input fails to keep the script running
    radius = 4
    print(f"Defaulting to radius = {radius}")

# --- CALCULATION SECTION ---
# Volume: V = π * r^2 * h
volume = math.pi * (radius**2) * height

# Total Surface Area: S = 2 * π * r * (r + h)
surface_area = 2 * math.pi * radius * (radius + height)

print("volume: ",volume)
print("surface_area: ",surface_area)

# EXTRA (2): Using f-strings with right-alignment and 2 decimal places
print("\n")
print("="*30)
print(f"Radius:{radius:>10.2f}")
#>10 this is the right-alignment, it reserves 10 spaces and pushes the number to the right edge
#.2 this rounds the number to exactly 2 decimal places
#f this ensures the number is displayed as standard decimal
print(f"Height:{height:>10.2f}")
print("-" * 30)
print(f"Volume:{volume:>10.2f}")
print(f"Surface Area:{surface_area:>10.2f}")
print("="*30)
````
